```python
def plot_format():
  plt.yscale('log')
  plt.xlabel('dimension of latent space')
  plt.ylabel('loss (MSE)')
  plt.legend(loc = 'upper right')
```

Activation functions:


```python
x = np.arange(-2,2,0.001)

l =  x

s = 1/(1 + np.exp(-x))

r = []
for value in x:
  r.append(max(value,0))

def tanh(x):
  return np.tanh(x)

plt.plot(x, l, label='linear')
plt.plot(x, s, label='sigmoid')
plt.plot(x, r, label='ReLu')
plt.plot(x, tanh(x), label='tanh')
plt.legend(loc='lower right', fontsize='x-large')
plt.ylim(-1.5,1.5)
plt.ylabel('\u03C8 (z)')
plt.xlabel('z')
plt.yticks(ticks=[-1,0,1])
plt.xticks(ticks=[-2,-1,0,1,2])
plt.grid()
plt.show()

```

```python
lst_dim = [1,2,4,8]

lst_Re = [20,30,40,50,60]
lst_AE = []
lst_POD = []

for i in range(4):
  sublst_AE = []
  sublst_POD = []
  for reynolds in AE_POD_dict:
    sublst_AE.append(AE_POD_dict[reynolds]['AE'][i])
    sublst_POD.append(AE_POD_dict[reynolds]['POD'][i])
  lst_AE.append(sublst_AE)
  lst_POD.append(sublst_POD)


fig, (ax1, ax2) = plt.subplots(1, 2)
fig.set_size_inches(15,5)

for i, dim in enumerate(lst_dim):
  ax1.plot(lst_Re, lst_AE[i], label='dim='+str(dim), marker='.')
ax1.set_yscale('log')
ax1.set_title('Auto-Encoder')
ax1.set_xlabel('Reynolds number')
ax1.set_ylabel('loss (MSE)')
ax1.set_xticks(lst_Re)
ax1.legend(loc='lower right')

for i, dim in enumerate(lst_dim):
  ax2.plot(lst_Re, lst_POD[i], '--', label='dim='+str(dim), marker='.')
ax2.set_yscale('log')
ax2.set_title('Proper Orthogonal Decomposition')
ax2.set_xlabel('Reynolds number')
ax2.set_ylabel('loss (MSE)')
ax2.set_xticks(lst_Re)
ax2.legend(loc='lower right')

plt.show()
```
